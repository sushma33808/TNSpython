Create a Student Report Card Program using:

1.List → store student names

2.Set → store subjects

3.Tuple → store student information (name, age)

4.Dictionary → store subject-wise marks

Requirements:

1.Ask the user to enter:

*Student name

*Age

*Number of subjects

2.For each subject:

*Enter subject name

*Enter marks

3.Store:

*Name in a list

*Subjects in a set

*Student info in a tuple

*Marks in a dictionary

4.Print a report card showing:

*Name

*Age

*Subjects

*Marks

*Total marks

*Average marks

```python
# List to store student names
student_names = []

# Set to store subjects
subjects = set()

# Dictionary to store subject-wise marks
marks = {}

# Get student details
name = input("Enter student name: ")
age = int(input("Enter age: "))
num_subjects = int(input("Enter number of subjects: "))

# Store name in list
student_names.append(name)

# Store student info in tuple
student_info = (name, age)

# Input subjects and marks
for i in range(num_subjects):
    subject = input("Enter subject name: ")
    mark = int(input("Enter marks: "))

    subjects.add(subject)      # Store subject in set
    marks[subject] = mark      # Store marks in dictionary

# Calculate total and average
total_marks = sum(marks.values())
average_marks = total_marks / num_subjects

# Display report card
print("\n----- STUDENT REPORT CARD -----")
print("Name :", student_info[0])
print("Age :", student_info[1])
print("Subjects :", subjects)
print("Marks :", marks)
print("Total Marks :", total_marks)
print("Average Marks :", average_marks)
```
