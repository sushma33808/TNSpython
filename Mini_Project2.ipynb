**problem: student marks analysis**
 1. create a dataset containing marks of 10 students
 2. students = ['A','B','C','D','E','F','G','H','I','J']
 3. marks = [78,85,92,67,88,95,72,81,90,76]

** TASKS: **
 1. Display student marks graphically
 2. Find the highest score
 3. Add appropriate title and axis labels
 4. Display marks in descending order and plot again

```python
import matplotlib.pyplot as plt

# Dataset
students = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
marks = [78, 85, 92, 67, 88, 95, 72, 81, 90, 76]

# Task 1: Display student marks graphically
plt.figure(figsize=(8,5))
plt.bar(students, marks)

# Task 3: Add title and axis labels
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()

# Task 2: Find the highest score
highest_score = max(marks)
print("Highest Score =", highest_score)

# Task 4: Display marks in descending order and plot again
sorted_data = sorted(zip(students, marks), key=lambda x: x[1], reverse=True)

sorted_students = []
sorted_marks = []

for student, mark in sorted_data:
    sorted_students.append(student)
    sorted_marks.append(mark)

print("\nMarks in Descending Order:")
for student, mark in sorted_data:
    print(student, ":", mark)

# Plot after sorting
plt.figure(figsize=(8,5))
plt.bar(sorted_students, sorted_marks)

plt.title("Student Marks Analysis (Descending Order)")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```
