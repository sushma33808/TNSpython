```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

Channels with the Highest Number of Subscribers

```python
top_subscribers=df[['Youtuber','subscribers']].sort_values(
    by='subscribers',
    ascending=False
).head(10)
print("Top 10 Channels by Subscribers")
print(top_subscribers)
plt.figure(figsize=(12,6))
sns.barplot(x='subscribers', y='Youtuber', data=top_subscribers, palette='viridis')
plt.title('Top 10 by Subscribers')
plt.xlabel('Subscribers')
plt.ylabel('youtuber')

plt.show()
```

```python
sns.lineplot(x="Youtuber",y="subscribers",data=df)
plt.show()
```

Which channels have the highest video views

```python
top_views=df[['Youtuber','video views']].sort_values(
    by='video views',
    ascending=False
).head(10)
print("Top 10 Channels by Video Views")
print(top_views)
plt.figure(figsize=(12,6))
sns.barplot(x='Youtuber', y='video views', data=top_views, palette='viridis')
plt.title('Top 10 by Subscribers')
plt.xlabel('youtuber')
plt.ylabel('Video views')
plt.show()
```

3. Countries with the Highest Number of YouTube Creators

```python
country_creators = df['Country'].value_counts().head(10)

print("Top Countries by Number of YouTube Creators")
print(country_creators)
plt.figure(figsize=(20, 6))
sns.barplot(x=country_creators.values, y=country_creators.index, palette='Blues_d')
plt.title('Top 10 Countries by Number of YouTube Creators')
plt.xlabel('Number of Creators')
plt.ylabel('Country')
plt.show()
```

```python
plt.pie(country_creators,data=country_creators,labels=country_creators.index,autopct='%1.1f%%')
plt.title("Highest number of Youtube Creators")
plt.show()
```
